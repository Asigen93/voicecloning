```python
from pydub import AudioSegment
import os

input_dir = "/content/dataset/mp3"
output_dir = "/content/dataset/wavs"
os.makedirs(output_dir, exist_ok=True)

for file in os.listdir(input_dir):
    if file.endswith(".mp3"):
        sound = AudioSegment.from_mp3(os.path.join(input_dir, file))
        file_name = os.path.splitext(file)[0] + ".wav"
        sound.export(os.path.join(output_dir, file_name), format="wav")

print("Semua file .mp3 sudah dikonversi ke .wav")
```

```python
import pandas as pd

df = pd.read_csv("/content/dataset/train.tsv", sep="\t")
df_tts = df[["path", "sentence"]].copy()

# Format sesuai LJSpeech: filename|kalimat|speaker
df_tts["tts_format"] = df_tts["path"].apply(lambda x: x.replace(".mp3", ".wav")) + "|" + df_tts["sentence"] + "|speakerX"

output_path = "/content/dataset/metadata.csv"
df_tts["tts_format"].to_csv(output_path, index=False, header=False, encoding="utf-8")

print(f"metadata.csv berhasil dibuat di {output_path}")
```

```python
!head -n 5 /content/dataset/metadata.csv
```

```python
import librosa, soundfile as sf
input_dir = "/content/dataset/wavs"
output_dir = "/content/dataset/wavs_proc"
os.makedirs(output_dir, exist_ok=True)

for file in os.listdir(input_dir):
    if file.endswith(".wav"):
        y, sr = librosa.load(os.path.join(input_dir, file), sr=22050, mono=True)
        y = y / max(abs(y))
        sf.write(os.path.join(output_dir, file), y, 22050)

print("Audio siap dipakai untuk training.")
```

```python
config_path = "/content/config.json"
config_json = """
{
  "model": "glow_tts",
  "run_name": "finetune_my_voice",
  "output_path": "/content/finetuned_model",
  "datasets": [
      {
        "formatter": "ljspeech",
        "path": "/content/dataset",
        "meta_file_train": "metadata.csv",
        "meta_file_val": "metadata.csv"
      }
  ],
  "trainer": {
      "epochs": 50,
      "batch_size": 16,
      "eval_batch_size": 16
  }
}
"""
with open(config_path, "w") as f:
    f.write(config_json)
```
