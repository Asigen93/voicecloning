```python
from pydub import AudioSegment
import os

input_dir = "/content/dataset/mp3"
output_dir = "/content/dataset/wavs"
os.makedirs(output_dir, exist_ok=True)

for file in os.listdir(input_dir):
    if file.endswith(".mp3"):
        sound = AudioSegment.from_mp3(os.path.join(input_dir, file))
        file_name = os.path.splitext(file)[0] + ".wav"
        sound.export(os.path.join(output_dir, file_name), format="wav")

print("Semua file .mp3 sudah dikonversi ke .wav")
```

```python
import pandas as pd

df = pd.read_csv("/content/dataset/train.tsv", sep="\t")
df_tts = df[["path", "sentence"]].copy()

# Format sesuai LJSpeech: filename|kalimat|speaker
df_tts["tts_format"] = df_tts["path"].apply(lambda x: x.replace(".mp3", ".wav")) + "|" + df_tts["sentence"] + "|speakerX"

output_path = "/content/dataset/metadata.csv"
df_tts["tts_format"].to_csv(output_path, index=False, header=False, encoding="utf-8")

print(f"metadata.csv berhasil dibuat di {output_path}")
```

```python
import librosa, soundfile as sf
input_dir = "/content/dataset/wavs"
output_dir = "/content/dataset/wavs_proc"
os.makedirs(output_dir, exist_ok=True)

for file in os.listdir(input_dir):
    if file.endswith(".wav"):
        y, sr = librosa.load(os.path.join(input_dir, file), sr=22050, mono=True)
        y = y / max(abs(y))
        sf.write(os.path.join(output_dir, file), y, 22050)

print("Audio siap dipakai untuk training.")
```

```python
import os
import re
import pandas as pd

# Path folder wavs dan metadata.csv
wavs_dir = "/content/dataset/wavs"
metadata_file = "/content/dataset/metadata.csv"
output_file = "/content/dataset/metadata_filtered.csv"

# --- 1. Ambil semua ID dari nama file audio di folder wavs ---
audio_files = [f for f in os.listdir(wavs_dir) if f.endswith(".wav")]

# Ekstrak pola id_xxxxx dari nama file
audio_ids = {re.search(r"id_\d+", f).group(0) for f in audio_files if re.search(r"id_\d+", f)}

print(f"Jumlah file audio di folder: {len(audio_files)}")
print(f"Contoh ID audio: {list(audio_ids)[:10]}")

# --- 2. Baca metadata.csv ---
df = pd.read_csv(metadata_file, sep="|", header=None, names=["filename", "text"])

# Ekstrak id_xxxxx dari kolom filename
df["id_key"] = df["filename"].apply(lambda x: re.search(r"id_\d+", str(x)).group(0) if re.search(r"id_\d+", str(x)) else None)

# --- 3. Filter metadata agar hanya yang ID-nya ada di folder wavs ---
df_filtered = df[df["id_key"].isin(audio_ids)]

# --- 4. Simpan hasil ---
df_filtered[["filename", "text"]].to_csv(output_file, sep="|", index=False, header=False)

print("✅ Metadata berhasil difilter!")
print(f"Jumlah baris awal: {len(df)}")
print(f"Jumlah baris akhir: {len(df_filtered)}")
print(f"Hasil disimpan di: {output_file}")
```
