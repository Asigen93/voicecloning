```python
from pydub import AudioSegment
import os
import warnings

# Hilangkan SyntaxWarning dari pydub
warnings.filterwarnings("ignore", category=SyntaxWarning)

input_dir = "/content/dataset/mp3"
output_dir = "/content/dataset/wavs"
os.makedirs(output_dir, exist_ok=True)

for file in os.listdir(input_dir):
    if file.endswith(".mp3"):
        sound = AudioSegment.from_mp3(os.path.join(input_dir, file))
        file_name = os.path.splitext(file)[0] + ".wav"
        sound.export(os.path.join(output_dir, file_name), format="wav")

print("✅ Semua file .mp3 sudah dikonversi ke .wav tanpa warning")
```

```python
import librosa, soundfile as sf
input_dir = "/content/dataset/wavs"
output_dir = "/content/dataset/wavs_proc"
os.makedirs(output_dir, exist_ok=True)

for file in os.listdir(input_dir):
    if file.endswith(".wav"):
        y, sr = librosa.load(os.path.join(input_dir, file), sr=22050, mono=True)
        y = y / max(abs(y))
        sf.write(os.path.join(output_dir, file), y, 22050)

print("Audio siap dipakai untuk training.")
```

```python
# ============================================
# Install dependensi
# ============================================
!pip install -q openai-whisper
!apt-get install -y ffmpeg

# ============================================
# Import library
# ============================================
import os
import whisper
import pandas as pd

# ============================================
# Path dataset
# ============================================
wavs_dir = "/content/dataset/wavs"
output_file = "/content/dataset/metadata_transcribed.csv"

# Load model Whisper (tiny lebih cepat, bisa ganti "base"/"small" kalau butuh lebih akurat)
model = whisper.load_model("tiny")

# ============================================
# Proses transkripsi semua file .wav
# ============================================
rows = []

for fname in sorted(os.listdir(wavs_dir)):
    if fname.endswith(".wav"):
        audio_path = os.path.join(wavs_dir, fname)
        # Transkripsi audio
        result = model.transcribe(audio_path, language="id")
        text = result["text"].strip()
        rows.append([fname, text])
        print(f"{fname} -> {text}")

# ============================================
# Simpan ke metadata.csv (format: filename|text)
# ============================================
df = pd.DataFrame(rows, columns=["filename", "text"])
df.to_csv(output_file, sep="|", index=False, header=False)

print(f"\n✅ metadata_transcribed.csv berhasil dibuat dengan {len(rows)} baris!")
print(f"Lokasi file: {output_file}")
```
